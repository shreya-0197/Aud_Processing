```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
sns.set(color_codes=True)
```

```python
df=pd.read_csv("audio_data.csv")
```

```python
df.columns
```

```python
n = 50

# Using DataFrame.tail() to Drop top n rows
df1 = df.tail(df.shape[0] -n)
print(df1)

```

```python
# Using drop() function to delete last n rows
n = 40
df1.drop(df1.tail(n).index,inplace = True)
print(df1)
```

```python
df2=df1.drop(['F-1', 'C0', 'F0', 'A0', 'C1', 'D1', 'F1', 'G1', 'A1', 'A♯1', 'C2',
       'C♯2', 'G6', 'G♯6',
       'A6', 'A♯6', 'B6', 'C7', 'C♯7', 'D7', 'D♯7', 'E7', 'F7', 'F♯7', 'G7',
       'G♯7', 'A7', 'A♯7', 'B7'], axis=1)
df2
```
