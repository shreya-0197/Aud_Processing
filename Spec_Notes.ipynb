```python
import numpy as np
import matplotlib.pyplot as plt
import librosa.display
import librosa

# Load audio file
audio_file = 'chopin.wav'
y, sr = librosa.load(audio_file)

# Compute spectrogram
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)

# Detect pitches or frequencies of notes
pitches, magnitudes = librosa.piptrack(y=y, sr=sr)

# Convert pitches to note names
def freq_to_note(frequency):
    note_num = int(round(12 * np.log2(frequency / 440.0))) + 69
    octave = (note_num // 12) - 1
    note_names = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']
    note_index = note_num % 12
    return f'{note_names[note_index]}{octave}'

# Overlay the notes on the spectrogram
plt.figure(figsize=(10, 6))
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')

for frame in range(pitches.shape[1]):
    for pitch, magnitude in zip(pitches[:, frame], magnitudes[:, frame]):
        if magnitude > 0:  # Threshold for considering a note
            freq = librosa.core.midi_to_hz(pitch)
            note = freq_to_note(freq)
            plt.text(librosa.frames_to_time(frame), freq, note, ha='center', fontsize=8, color='red')

plt.title('Spectrogram with detected notes')
#plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa.display
import librosa

# Load audio file
audio_file = 'chopin.wav'
y, sr = librosa.load(audio_file)

# Compute spectrogram
D = np.abs(librosa.stft(y))

# Convert magnitude spectrogram to dB scale
D_db = librosa.amplitude_to_db(D, ref=np.max)

# Compute pitches
pitches, magnitudes = librosa.piptrack(y=y, sr=sr)

# Convert pitches to notes
def freq_to_note(frequency):
    note_num = 12 * (np.log2(frequency) - np.log2(440)) + 69
    return int(round(note_num))  # MIDI note number

# Overlay notes on spectrogram
plt.figure(figsize=(12, 8))
librosa.display.specshow(D_db, sr=sr, x_axis='time', y_axis='hz')

for frame in range(pitches.shape[1]):
    for freq, magnitude in zip(pitches[:, frame], magnitudes[:, frame]):
        if np.isfinite(freq) and freq > 0:  # Check if frequency is valid and positive
            note = freq_to_note(freq)
            plt.text(frame * 512 / sr, freq, librosa.midi_to_note(note), color='red')

plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram with Notes')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.tight_layout()
plt.show()
```
